```python
from nba_api.stats.endpoints import teamgamelogs
teamgamelog = teamgamelogs.TeamGameLogs()
teamgamelog_df = teamgamelog.get_data_frames()[0]
#teamgamelog_df
teamgamelog_sorted_df = teamgamelog_df.sort_values(by=['TEAM_ABBREVIATION'], ascending=True).dropna()
teamgamelog_sorted_df
```

```python
team_dict={"ATL": [0,0]}
for ind in teamgamelog_sorted_df.index:
    #print(teamgamelog_sorted_df['TEAM_ABBREVIATION'][ind],teamgamelog_sorted_df['WL'][ind])
    name = teamgamelog_sorted_df['TEAM_ABBREVIATION'][ind]
    if(name not in team_dict):
        team_dict[name]=[0,0]
    if(teamgamelog_sorted_df['WL'][ind]=="W"):
        team_dict[name][0]+=1
    else:
        team_dict[name][1]+=1
    

#print(type(team_dict))
team_dict
```

```python
lal = [team for team in team_dict if team['full_name']=='Los Angeles Lakers'][0]
lal_id = lal['id']
lal_id
```

```python
from nba_api.stats.endpoints import playergamelog
from nba_api.stats.library.parameters import SeasonAll
gamelog_bron = playergamelog.PlayerGameLog(player_id = '2544', season = '2017')
gamelog_bron_df = gamelog_bron.get_data_frames()[0]

gamelog_bron_all = playergamelog.PlayerGameLog(player_id = '2544', season = SeasonAll.all)
gamelog_bron_df_all = gamelog_bron_all.get_data_frames()[0]
gamelog_bron_df_all
```

```python
from nba_api.stats.endpoints import leaguegamefinder
lal_games = leaguegamefinder.LeagueGameFinder(team_id_nullable=lal_id).get_data_frames()[0]
lal_games
```

```python
win_nums = 0
lose_nums = 0
for games in lal_games["WL"]:
    if games == "W": 
        win_nums += 1
    else:
        lose_nums += 1
print("number of winning games = ", win_nums, "number of losing games = ", lose_nums)
print("win/lose ratio = ", win_nums/lose_nums)
```
